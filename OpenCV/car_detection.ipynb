```python
import numpy as np
import cv2 as cv
```

```python
haar_cascade_car_path = './cascades/haarcascade_car.xml'
```

```python
car_cascade = cv.CascadeClassifier(haar_cascade_car_path)
```

```python
video = cv.VideoCapture('videos/Bris.mp4')
```

```python
while video.isOpened():
    ret, frame = video.read()
    if ret:
        # print('Success')
        frame_gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
        cars = car_cascade.detectMultiScale(frame_gray, 1.2, 3) # scale factor, minimum neighbors
            # draw rectangles and display
        for (x,y, w, h) in cars:
            cv.rectangle(frame, (x,y), (x+w, y+h), (0, 128, 0), 2) # color and thickness
            cv.imshow('Cars', frame)
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
video.release()
cv.destroyAllWindows()
```
