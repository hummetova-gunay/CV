```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
waldo = cv.imread('./images/waldo.jpg')
waldo_beach = cv.imread('./images/WaldoBeach.jpg')
```

```python
waldo.shape, waldo_beach.shape
```

```python
height, width = waldo.shape[:2]
height, width
```

```python
waldo_gray = cv.cvtColor(waldo, cv.COLOR_BGR2GRAY)
waldo_beach_gray = cv.cvtColor(waldo_beach, cv.COLOR_BGR2GRAY)
```

```python
match = cv.matchTemplate(waldo_beach_gray, waldo_gray, cv.TM_CCOEFF)
match.shape
```

```python
min_score, max_score, min_loc, max_loc = cv.minMaxLoc(match, mask= None)
```

```python
max_loc # top left corner
```

```python
bottom_right = (max_loc[0] + width, max_loc[1]+height)
```

```python
bottom_right
```

```python
waldo_beach_matched = cv.rectangle(waldo_beach, max_loc, bottom_right, (0,0,255), 5)
```

```python
plt.imshow(cv.cvtColor(waldo_beach_matched, cv.COLOR_BGR2RGB))
```
