```python
import os
cascade_path = cv.data.haarcascades  # OpenCV 4.x provides this built-in
print(cascade_path)
```

```python
frontalface_cascade = cv.CascadeClassifier(
    cv.data.haarcascades + 'haarcascade_frontalface_default.xml'
)
smile_cascade = cv.CascadeClassifier(
    cv.data.haarcascades + 'haarcascade_smile.xml'
)
```

```python
print(frontalface_cascade.empty())  # True = failed to load
print(smile_cascade.empty())        # True = failed to load
```

```python
capture = cv.VideoCapture(0)

while True:
    res, frame = capture.read()
    frame_gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

    # detect faces
    faces = frontalface_cascade.detectMultiScale(frame_gray, 1.3, 5)
    for (x,y,w,h) in faces:
        cv.rectangle(frame, (x,y), (x+w, y+h), (0, 250, 0), 2)
        cropped_gray = frame_gray[y:y+h, x:x+w]
        cropped_color = frame[y:y+h, x:x+w]
        smiles = smile_cascade.detectMultiScale(cropped_gray, 1.6, 22)
        for (sx, sy, sw, sh) in smiles:
            cv.rectangle(cropped_color, (sx, sy), (sx + sw, sy + sh), (0, 0, 255), 2)
    cv.imshow('Smile Detector', frame)    
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

capture.release()
cv.destroyAllWindows()
```
